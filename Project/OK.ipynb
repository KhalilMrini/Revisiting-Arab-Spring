```python
import pickle
import numpy as np
import pandas as pd
import os
```

```python
AR = 'ar'
EN = 'en'
FR = 'fr'
LANGS = [AR, EN, FR]

TN = 'Tunisia'
EG = 'Egypt'
COUNTRS = [TN, EG]

SM = 'SOCIAL_MEDIA'
WB = 'WEBLOG'
CL = 'CLASSIFIED'
FM = 'FORUM'
MN = 'MAINSTREAM_NEWS'
TYPES = [SM, WB, CL, FM, MN]
```

```python
STATS = {}
for country in COUNTRS:
    dict_countr = {}
    for lang in LANGS:
        dict_lang = {}
        for t in TYPES:
            dict_lang[t] = 0
        dict_countr[lang] = dict_lang
    STATS[country] = dict_countr

directory = 'Desktop/data/data.pkl/dataframes/'
```

```python
total = 0
for filename in os.listdir(directory):
    df = pickle.load(open(os.path.join(directory, filename), 'rb'))
    df['Country'] = np.sign(df.Egypt- df.Tunisia)
    df[EG] = df.Country >= 0
    df[TN] = df.Country <=0
    for country in COUNTRS:
        for lang in LANGS:
            for typ in TYPES:
                STATS[country][lang][typ] += len(df[(df[country]) & (df.Lang_Code == lang) & (df.Type == typ)])        
    total += len(df)
    print('The dataframe {} contains {} items'.format(filename, len(df)))
print('Total number of elements {}'.format(total))
```

```python
pickle.dump(STATS, open('stats_dict.pkl', 'wb'))
```

### Clean Contents

```python
from os import listdir
from bs4 import BeautifulSoup
def to_text(content):
    try:
        return BeautifulSoup(content, 'lxml').text
    except:
        return content
```

```python
directory = 'Desktop/data/data.pkl/dataframes/'
i = 0
for filename in os.listdir(directory):
    if filename.endswith('.pkl'):
        print(filename)
        i += 1
        print('Loading Dataframe{} ... '.format(i))
        df = pickle.load(open(os.path.join(directory, filename), 'rb'))
        print('The dataframe {} contains {} items'.format(filename, len(df)))
        print('Cleaning Dataframe{} ... '.format(i))
        df.Post_Content = df.Post_Content.apply(to_text)
        df.to_pickle('clean_{}.pkl'.format(i))
print('Total number of elements {}'.format(total))
```
